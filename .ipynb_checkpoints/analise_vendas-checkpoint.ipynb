```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
import numpy as np

df = pd.read_csv('vendas.csv')

# Definindo estilo do gráfico
sns.set_theme()
```

```python
# Total de vendas por cidade
vendas_por_cidade = df.groupby('Cidade')['Valor Total'].sum().sort_values()

# Gráfico de barras
plt.figure(figsize=(10, 6))
vendas_por_cidade.plot(kind='barh', color='skyblue')
plt.title('Todal de Vendas por Cidade')
plt.xlabel('Valor Total (R$)')
plt.ylabel('Cidade')
plt.grid(axis='x')
plt.show()
```

```python
# Total de vendas por Categoria
vendas_por_categoria = df.groupby('Categoria')['Valor Total'].sum()

# Gráfico de Pizza
plt.figure(figsize=(8, 8))
vendas_por_categoria.plot(kind='pie', autopct='%1.1f%%', startangle=140, color=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribuição de vendas por Categoria')
plt.ylabel('')
plt.show()
```

```python
# Total de vendas por canal de vendas
vendas_por_canal = df.groupby('Canal de Venda')['Valor Total'].sum()

# Gráfico de Barras
plt.figure(figsize=(8, 6))
vendas_por_canal.plot(kind='bar', color='coral')
plt.title('Total de vendas por Canal de Venda')
plt.xlabel('Canal de Venda')
plt.ylabel('Valor Total (R$)')
plt.grid(axis='y')
plt.show()
```

```python
# Vendas ao longo do tempo
df['Data'] = pd.to_datetime(df['Data'])

# Agrupar por Mês e somar as vendas
vendas_mensais = df.groupby(df['Data'].dt.to_period('M'))['Valor Total'].sum()

# Gráfico de linha
plt.figure(figsize=(12, 6))
vendas_mensais.plot(kind='line', marker='o', linestyle='-', color='blue')
plt.title('Vendas ao Longo do Tempo')
plt.xlabel('Mês')
plt.ylabel('Valor Total (R$)')
plt.grid(axis='both')
plt.show()
```

```python
# Utilizando o Streamlit para criação dos gráficos

# Função para carregar os dados
@st.cache
def load_data():
    df = pd.read_csv('vendas.csv')
    df['Data'] = pd.to_datatime(df['Data'])
    return df

# Carregar os dados
df = load_data()

# Título da aplicação
st.title('Análise de Vendas de Produtos Brasileiros')

# Extrair dataset
st.subheader('Dados de Vendas')
st.write(df.head())

# Análise 1: Total de Vendas por Cidade
st.subheader('Total de Vendas por Cidade')
vendas_pro_cidade = df.groupby('Cidade')['Valor Total'].sum().sort_values()
fig, ax = plt.subplots(figsize=(10, 6))
vendas_por_cidade.plot(kind='barh', color='skyblue', ax=ax)
ax.set_title('Total de Vendas por Cidade')
ax.set_xlabel('Valor Total (R$)')
ax.set_ylabel('Cidade')
ax.grid(axis='x')
st.pyplot(fig)

# Análise 2: Total de Vendas por Categoria
st.subheader('Distribuição de Vendas por Categoria')
vendas_por_categoria = df.grpupby('Categoria')['Valor Total'].sum()
fig, ax = plt.subplots(figsize=(8, 8))
vendas_por_categoria.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], ax=ax)
ax.set_title('Distribuição de Vendas por Categoria')
ax.set_ylabel('')
st.pyplot(fig)

# Análise 3: Total de Vendas por Canal de Vendas
st.subheader('Total de Vendas por Canal de Vendas')
vendas_por_canal = df.groupby('Canal de Venda')['Valor Total'].sum()
fig, ax = plt.subplots(figsize=(8, 6))
vendas_por_canal.plot(kind='bar', color='coral', ax=ax)
ax.set_title('Total de Vendas por Canal de Venda')
ax.set_xlabel('Canal de Venda')
ax.set_ylabel('Valor Total (R$)')
ax.grid(axis='y')
st.pyplot(fig)

# Análise 4: Vendas ao Longo do Tempo
st.subheader('Vendas ao Longo do Tempo')
vendas_mensais = df.groupby(df['Data'].dt.to_period('M'))['Valor Toal'].sum()
fig, ax = plt.subplots(figsize=(12, 6))
vendas_mensais.plot(kind='line', marker='o', linestyle='-', color='blue', ax=ax)
ax.set_title('Vendas ao Longo do Tempo')
ax.set_xlabel('Mês')
ax.set_ylabel('Valor Total (R$)')
ax.grid(axis='both')
st.pyplot(fig)

# Análise 5: Top Produtos Vendidos
st.subheader('Top Produtos Vendidos')
top_produtos = df.groupby('Protudo')['Valor Total'].sum().sort_values(ascending=False).head(10)
fig, ax = plt.subplots(figsize=(10, 6))
top_produtos.plot(kind='bar', color='purple', ax=ax)
ax.set_title('Top 10 Protudos Vendidos')
ax.set_xlabel('Protudo')
ax.set_ylabel('Valor Total (R$)')
ax.grid(axis='y')
st.pyplot(fig)
```
